# Intro to pandas
The two main data structures in Pandas are the Series and DataFrame

Series are like columns and Dataframes are whole table

# Series or 1D

```python
# pandas series
a = [1,7,2]
a1 = pd.Series(a)
a1
```

```python
# labeling - used to access value
a = [1,7,2]
a1 = pd.Series(a)
a1[0]
```

```python
# with create label we can create our own labels
a = [1,7,2]
a1 = pd.Series(a,index=["x","y","z"])
a1
```

```python
a = [1,7,2]
a1 = pd.Series(a,index=["x","y","z"])
a1["x"]
```

```python
# we can also use a key or value object like dictionary, when creating a series

cal = {"day1":420,"day2":380,"day3":390}
a = pd.Series(cal)
a
```

```python
# now we will create a series using only data from day1 and day2

cal = {"day1":420,"day2":380,"day3":390}
result = pd.Series(cal, index=["day1","day2"])
result
```

# Dataframe or 2D
usually multidimensional, contains rows and columns both like an excel sheet, created with 2 series

```python
a = {"cal":[420,380,390],"duration":[50,40,45]}
a1 = pd.DataFrame(a)
a1
```

```python
# access using location attribute

a = {"cal":[420,380,390],"duration":[50,40,45]}
a1 = pd.DataFrame(a)
a1.loc[0]
```

```python
a = {"cal":[420,380,390],"duration":[50,40,45]}
a1 = pd.DataFrame(a)
a1.loc[[0,1]] # use 2D bracket
```

# named index
with the index arg, we can name our own indexin Dataframe

```python
a = {"cal":[420,380,390],"duration":[50,40,45]}
a1 = pd.DataFrame(a, index=["day1","day2","day3"])
a1
```

```python
# locate using the named indexes

a = {"cal":[420,380,390],"duration":[50,40,45]}
a1 = pd.DataFrame(a, index=["day1","day2","day3"])
a1.loc["day1"] # output is in series form
```

```python
a = {"cal":[420,380,390],"duration":[50,40,45]}
a1 = pd.DataFrame(a, index=["day1","day2","day3"])
a1.loc[["day1"]] # output is in dataframe form
```

# Data cleaning
Bad data could be- duplicate data, wrong data, emplty cells, wrong format

```python
df = pd.read_csv('/content/dirtydata.csv')
print(df.to_string())
```

For Empty Cells

```python
# Drop the cell
cleaned_data = df.dropna()
print(cleaned_data.to_string())
```

```python
cleaned_data.info()
```

```python
"""
If at any case we want to make changes in the original dataset then use
'inplace=True'.
It will remove the rows containing the null values.
df.dropna(inplace=True)
"""
```

```python
# Replace empty values
"""
cleaned_data = df.fillna(130,inplace=True)
print(cleaned_data.to_string())
"""

# Replace empty value at a particular place,add the column name
"""
cleaned_data = df["Calories"].fillna(130,inplace=True)
print(cleaned_data.to_string())
"""
```

```python
# Here we can also make replacements using mean(), median() or mode()

# Mean: Calculate avg value and fill w/ mean
x = cleaned_data["Calories"].mean()
cleaned_data["Calories"].fillna(x, inplace=True)
print(cleaned_data.to_string())
```

```python
# Median
x_median = cleaned_data["Calories"].median()
cleaned_data["Calories"].fillna(x_median, inplace=True)
print(cleaned_data.to_string())
```

```python
# Mode
x_mode = cleaned_data["Calories"].mode()[0]
cleaned_data["Calories"].fillna(x_mode, inplace=True)
print(cleaned_data.to_string())
```
